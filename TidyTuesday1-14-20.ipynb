```python
import numpy as np
import pandas as pd
```

```python
passwords = pd.read_csv("passwords.csv")
passwords.head()
```

```python
passwords.corr()
```

```python
pd.get_dummies(passwords, columns=['category'])
```

```python
passwords.isnull()
```

```python
passwords.dropna(inplace=True)
```

```python
passwords
```

```python
print(passwords.corr())

X = passwords.drop(['strength', 'category', 'time_unit','password'], axis=1)
y = passwords[['strength']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
```

```python
reg = LinearRegression()
rid = Ridge()
Las = Lasso()
en = ElasticNet()
```

```python
reg.fit(X_train, y_train)
rid.fit(X_train, y_train)
Las.fit(X_train, y_train)
en.fit(X_train, y_train)
```

```python
reg.coef_
```

```python
reg.predict(X_test)
reg.score(X_test, y_test)
```

```python
rid.score(X_test, y_test)
```

```python
Las.score(X_test, y_test)
```

```python
en.score(X_test, y_test)
```
