```python
COUNTRY = data['country'].unique()
PARENT_COMPANY = data['parent_company'].unique()
YEARS = data['year'].unique()
```

```python
data2.groupby(by=['parent_company'],as_index=False).mean()
```

```python
year_diff_dict = {}
for company in PARENT_COMPANY:
    #ic(company)
    #ic((data2['grand_total'][(data2['parent_company']==company)&(data2['year']==2019)].values[0]))
    #ic(data2['grand_total'][(data2['parent_company']==company)&(data2['year']==2020)].values[0])
    try:
        year_diff_dict[company] = data2['grand_total'][(data2['parent_company']==company) & (data2['year']==2020)].values[0] - data2['grand_total'][(data2['parent_company']==company) & (data2['year']==2019)].values[0] 
    except:
        pass
```

```python
year_diff_dict
```

```python
years_diff_list = sorted(year_diff_dict.values())
sorted_years = {}
for i in years_diff_list:
    for k in year_diff_dict.keys():
        if year_diff_dict[k] ==i:
            sorted_years[k] = year_diff_dict[k]
print(sorted_years)
```
